```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
imagem_base = (cv2.imread('/content/drive/MyDrive/bleh.png'))
cv2_imshow(imagem_base)
print(imagem_base.shape)
```

```python
# B = 0, G = 1, R = 2
imagemR = imagem_base.copy()
imagemR[:,:,0] = 0
imagemR[:,:,1] = 0
cv2_imshow(imagemR)
```

```python
imagemG = imagem_base.copy()
imagemG[:,:,0] = 0
imagemG[:,:,2] = 0
cv2_imshow(imagemG)
```

```python
imagemB = imagem_base.copy()
imagemB[:,:,1] = 0
imagemB[:,:,2] = 0
cv2_imshow(imagemB)
```

```python
imagem_somaRG = imagemR + imagemG
cv2_imshow(imagem_somaRG)
```

```python
imagem_somaGB = imagemG + imagemB
cv2_imshow(imagem_somaGB)
```

```python
imagem_somaRB = imagemR + imagemB
cv2_imshow(imagem_somaRB)
```

```python
imagem_somaRGB = imagemR + imagemG + imagemB
cv2_imshow(imagem_somaRGB)
```
