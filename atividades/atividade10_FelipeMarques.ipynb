```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = np.array(cv2.imread('/content/drive/MyDrive/PDI/test-pattern.png', 0))
cv2_imshow(img)
```

```python
def criarKernel(m, n):
 kernel = np.ones((m,n))
 kernel = kernel/kernel.sum()
 return kernel


def kernelGau(m, n):
 kg = np.zeros((m,n))
 return kg
```

```python
m = 3
n = 3
k3 = criarKernel(m,n)


a = int((m-1)/2)
b = int((n-1)/2)

M = img.shape[0]  
N = img.shape[1] 

def padding (a, b, M, N):

 imgPadded = np.zeros((M+2*a, N+2*b))
 imgPadded[a:(M+a),b:(N+b)] = img
 return imgPadded

imgP = padding(a, b, M, N) 


imgFiltrada = np.zeros(imgP.shape)
for i in range(a, a+M):
  for j in range(b, b+N):
    vizinhanca = imgP[i-a:i+a+1, j-b:j+b+1]
    imgFiltrada[i,j] = np.sum(vizinhanca*k3)



imgFiltrada = imgFiltrada[a:a+M, b:b+M]
print('Resultado da Função Box - 3x3')
cv2_imshow(np.hstack((img, imgFiltrada)))
```

```python
m=3
n=3

g3=kernelGau(m, n)
sigma = 1

for i in range(g3.shape[0]):
  for j in range(g3.shape[1]):
    g3[i,j] = np.exp(-((i-(m-1)/2)**2 + (j-(n-1)/2)**2)/sigma**2)


g3 = g3/g3.sum()

print(g3)


a = int((m-1)/2)
b = int((n-1)/2)

M = img.shape[0]  
N = img.shape[1] 

def padding (a, b, M, N):

 imgPadded = np.zeros((M+2*a, N+2*b))
 imgPadded[a:(M+a),b:(N+b)] = img
 return imgPadded

imgP = padding(a, b, M, N) 


imgFiltradag3 = np.zeros(imgP.shape)
for i in range(a, a+M):
  for j in range(b, b+N):
    vizinhanca = imgP[i-a:i+a+1, j-b:j+b+1]
    imgFiltradag3[i,j] = np.sum(vizinhanca*g3)


imgFiltradag3 = imgFiltradag3[a:a+M, b:b+M]
print('Resultado da Função Gaussiano - 3x3')
cv2_imshow(np.hstack((img, imgFiltradag3)))
```
