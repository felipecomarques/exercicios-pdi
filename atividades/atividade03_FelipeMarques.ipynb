```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
imagem1 = np.array(cv2.imread('drive/MyDrive/PDI/lena_noise1.png', cv2.IMREAD_GRAYSCALE)) / 10
imagem2 = np.array(cv2.imread('drive/MyDrive/PDI/lena_noise2.png', cv2.IMREAD_GRAYSCALE)) / 10
imagem3 = np.array(cv2.imread('drive/MyDrive/PDI/lena_noise3.png', cv2.IMREAD_GRAYSCALE)) / 10
imagem4 = np.array(cv2.imread('drive/MyDrive/PDI/lena_noise4.png', cv2.IMREAD_GRAYSCALE)) / 10
imagem5 = np.array(cv2.imread('drive/MyDrive/PDI/lena_noise5.png', cv2.IMREAD_GRAYSCALE)) / 10
imagem6 = np.array(cv2.imread('drive/MyDrive/PDI/lena_noise6.png', cv2.IMREAD_GRAYSCALE)) / 10
imagem7 = np.array(cv2.imread('drive/MyDrive/PDI/lena_noise7.png', cv2.IMREAD_GRAYSCALE)) / 10
imagem8 = np.array(cv2.imread('drive/MyDrive/PDI/lena_noise8.png', cv2.IMREAD_GRAYSCALE)) / 10
imagem9 = np.array(cv2.imread('drive/MyDrive/PDI/lena_noise9.png', cv2.IMREAD_GRAYSCALE)) / 10
imagem10 = np.array(cv2.imread('drive/MyDrive/PDI/lena_noise10.png', cv2.IMREAD_GRAYSCALE)) / 10
soma = (imagem1 + imagem2 + imagem3 + imagem4 + imagem5 + imagem6 + imagem7 + imagem8 + imagem9 + imagem10)
```

```python
cv2_imshow(soma)
print (soma)
```

```python
shader = np.array(cv2.imread('drive/MyDrive/PDI/shader.png', cv2.IMREAD_GRAYSCALE))
final = soma * shader
final = final / final.max()
final = final * 255
```
