```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def funcaoHistograma (img, indiceImagem, titulo):
  plt.figure(indiceImagem)
  img_bins = plt.hist(img.flatten(), bins = 256, range = (-0.5, 255.5), density =True)
  plt.xlabel("Níveis de Intensidade")
  plt.ylabel("Frequência")
  plt.title(titulo)
  plt.show()
```

```python
cat = cv2.imread('/content/drive/MyDrive/PDI/cat.jpeg', 0)
catEqualizado = cv2.equalizeHist(cat)

cv2_imshow(cat)
funcaoHistograma(cat, 1, "Imagem do gato sem equalização")
cv2_imshow(catEqualizado)
funcaoHistograma(catEqualizado, 2, "Imagem do gato com equalização")
```

```python
butterfly = cv2.imread('/content/drive/MyDrive/PDI/butterfly.jpg', 0)
butterflyEqualizado = cv2.equalizeHist(butterfly)

cv2_imshow(butterfly)
funcaoHistograma(butterfly, 3, "Imagem da borboleta sem equalização")
cv2_imshow(butterflyEqualizado)
funcaoHistograma(butterflyEqualizado, 4, "Imagem da borboleta sem equalização")
```

```python
hawkesbay = cv2.imread('/content/drive/MyDrive/PDI/hawkes_bay_NZ.jpg', 0)
hawkesbayEqualizado = cv2.equalizeHist(hawkesbay)

cv2_imshow(hawkesbay)
funcaoHistograma(hawkesbay, 5, "Imagem da hawkesbay sem equalização")
cv2_imshow(hawkesbayEqualizado)
funcaoHistograma(hawkesbayEqualizado, 6, "Imagem da hawkesbay com equalização")
```
