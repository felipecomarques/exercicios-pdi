```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
cidade = np.array(cv2.imread('/content/drive/MyDrive/PDI/aerial_city_view.png'))
cv2_imshow(cidade)

rhino = np.array(cv2.imread('/content/drive/MyDrive/PDI/rhino.png'))
cv2_imshow(rhino)
```

```python
plt.figure(1)
cidade_pdf = plt.hist(cidade.flatten(), bins = 256, range = (-0.5, 255.5), density =True)
plt.xlabel("Nveis de Intensidade")
plt.ylabel("Frequncia")
plt.title("Histograma de cidade.png")
plt.show()
```

```python
plt.figure(2)
rhino_pdf = plt.hist(rhino.flatten(), bins = 256, range = (-0.5, 255.5), density =True)
plt.xlabel("Nveis de Intensidade")
plt.ylabel("Frequncia")
plt.title("Histograma de rhino.png")
plt.show()
```

```python
def Tgamma (img, gamma, epsilon = 0):
  img_t = img.copy()
  c = 255 / ((img.max() + epsilon) ** gamma)
  img_t = c*((img + epsilon) **gamma)
  img_t = np.around(img_t)
  return img_t

cidade_gamma = Tgamma(cidade, 2.0)
cv2_imshow(cidade_gamma)

rhino_gamma = Tgamma(rhino, 0.5)
cv2_imshow(rhino_gamma)
```

```python
plt.figure(3)
cidade_gamma_pdf = plt.hist(cidade_gamma.flatten(), bins = 256, range = (-0.5, 255.5), density =True)
plt.xlabel("Nveis de Intensidade")
plt.ylabel("Frequncia")
plt.title("Histograma de cidade_gamma.png")
plt.show()
```

```python
plt.figure(4)
rhino_gamma_pdf = plt.hist(rhino_gamma.flatten(), bins = 256, range = (-0.5, 255.5), density =True)
plt.xlabel("Nveis de Intensidade")
plt.ylabel("Frequncia")
plt.title("Histograma de rhino_gamma.png")
plt.show()
```
