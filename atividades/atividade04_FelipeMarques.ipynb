```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
A = np.array(cv2.imread('drive/MyDrive/bleh.png', cv2.IMREAD_GRAYSCALE))
cv2_imshow(A)
```

```python
A_neg = 255 - A
cv2.imwrite('drive/MyDrive/bleh_negativo.png', A_neg)
cv2_imshow(A_neg)
```

```python
A_diminuir = np.array([[0.5, 0, 0],
                       [0, 0.5, 0],
                       [0, 0, 1]], dtype = 'f')
print (A_diminuir)
```

```python
A_T1 = cv2.warpPerspective(A, A_diminuir, (int(A.shape[1]*0.5),int(A.shape[0]*0.5)))
cv2_imshow(A_T1)
print (A_T1)
```

```python
A_aumentar = np.array([[1.5, 0, 0],
                       [0, 1.5, 0],
                       [0, 0, 1]], dtype = 'f')
print (A_aumentar)
```

```python
A_T2 = cv2.warpPerspective(A, A_aumentar, (int(A.shape[1]*1.5),int(A.shape[0]*1.5)))
cv2_imshow(A_T2)
print (A_T2)
```
