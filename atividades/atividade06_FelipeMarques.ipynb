```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
rhino = np.array(cv2.imread('/content/drive/MyDrive/PDI/rhino.png'))
cv2_imshow(rhino)
```

```python
def Tgamma (img, gamma, epsilon = 0):
  img_t = img.copy()
  c = 255 / ((img.max() + epsilon) ** gamma)
  img_t = c*((img + epsilon) **gamma)
  img_t = np.around(img_t)
  return img_t

rhino_gamma = Tgamma(rhino, 0.5)
cv2_imshow(rhino_gamma)
```

```python
aviao = np.array(cv2.imread('/content/drive/MyDrive/PDI/plane_sky.png'))
cv2_imshow(aviao)
```

```python
def Thresholding (img, k):
  img_t = img.copy()
  img_t[img_t > k] = 255
  img_t[img_t <= k] = 0
  return img_t

aviao_thrs = Thresholding(aviao, 146)
cv2_imshow(aviao_thrs)
```

```python
arara = np.array(cv2.imread('/content/drive/MyDrive/PDI/arara.png'))
cv2_imshow(arara)
```

```python
def Tcores(img, k):
  img_t = img.copy()
  for i in range(k):
    img_t[(img_t >= 256/k*i) & (img_t < 256/k*(i+1))] = 256/k*i + 256/(2*k)
  return img_t

arara_8cores = Tcores(arara, 8)
cv2_imshow(arara_8cores)
```
